```python
dt_underemploy_gender.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_underemploy_gender.columns.values[0] = 'Giới tính'

dt_underemploy_academic.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_underemploy_academic.columns.values[0] = 'Học vấn'

dt_underemploy_career.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_underemploy_career.columns.values[0] = 'Ngành'

dt_underemploy_region.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_underemploy_region.columns.values[0] = 'Giới tính'
dt_underemploy_region.columns.values[1] = 'Vùng'

dt_underemploy_age.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_underemploy_age.columns.values[0] = 'Nhóm tuổi'

# thất nghiệp
dt_unemploy_region.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_unemploy_region.columns.values[0] = 'Giới tính'
dt_unemploy_region.columns.values[1] = 'Vùng'

dt_unemploy_gender.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_unemploy_gender.columns.values[0] = 'Giới tính'

dt_unemploy_age.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_unemploy_age.columns.values[0] = 'Nhóm tuổi'

dt_unemploy_academic.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_unemploy_academic.columns.values[0] = 'Học vấn'

dt_birth.rename(columns={'Sơ bộ 2022': '2022'}, inplace=True)
dt_birth.columns.values[0] = 'Vùng'
```

```python
# Melt dữ liệu từ cả hai DataFrame
dt_unemploy_region = dt_unemploy_region.melt(id_vars=['Vùng','Giới tính'], var_name='Năm', value_name='Tỷ lệ thất nghiệp')
dt_underemploy_region = dt_underemploy_region.melt(id_vars=['Vùng','Giới tính'], var_name='Năm', value_name='Tỷ lệ thiếu việc làm')

# Kết hợp hai DataFrame theo cột 'Giới tính' và 'Năm'
combined_region = pd.merge(dt_unemploy_region, dt_underemploy_region, on=['Vùng', 'Năm','Giới tính'])
combined_region['Năm'] = combined_region['Năm'].astype('int64')
combined_region.head()
```

```python
dt_unemploy_academic = dt_unemploy_academic.melt(id_vars=['Học vấn'], var_name='Năm', value_name='Tỷ lệ thất nghiệp')
dt_underemploy_academic = dt_underemploy_academic.melt(id_vars=['Học vấn'], var_name='Năm', value_name='Tỷ lệ thiếu việc làm')

# Kết hợp hai DataFrame theo cột 'Giới tính' và 'Năm'
combined_academic = pd.merge(dt_unemploy_academic, dt_underemploy_academic, on=['Học vấn', 'Năm'])
combined_academic['Năm'] = combined_academic['Năm'].astype('int64')
combined_academic.head()
```

```python
# Melt dữ liệu từ cả hai DataFrame
dt_unemploy_age = dt_unemploy_age.melt(id_vars=['Nhóm tuổi'], var_name='Năm', value_name='Tỷ lệ thất nghiệp')
dt_underemploy_age = dt_underemploy_age.melt(id_vars=['Nhóm tuổi'], var_name='Năm', value_name='Tỷ lệ thiếu việc làm')

# Kết hợp hai DataFrame theo cột 'Giới tính' và 'Năm'
combined_age = pd.merge(dt_unemploy_age, dt_underemploy_age, on=['Nhóm tuổi', 'Năm'])
combined_age['Năm'] = combined_age['Năm'].astype('int64')
combined_age.head()
```

```python
dt_unemploy_gender = dt_unemploy_gender.melt(id_vars=['Giới tính'], var_name='Năm', value_name='Tỷ lệ thất nghiệp')
dt_underemploy_gender = dt_underemploy_gender.melt(id_vars=['Giới tính'], var_name='Năm', value_name='Tỷ lệ thiếu việc làm')

# Kết hợp hai DataFrame theo cột 'Giới tính' và 'Năm'
combined_gender = pd.merge(dt_unemploy_gender, dt_underemploy_gender, on=['Giới tính', 'Năm'])
combined_gender['Năm'] = combined_gender['Năm'].astype('int64')
combined_gender
```

```python
dt_underemploy_career = dt_underemploy_career.melt(id_vars=['Ngành'], var_name='Năm', value_name='Tỷ lệ thiếu việc làm')
dt_underemploy_career['Năm'] = dt_underemploy_career['Năm'].astype('int64')
dt_underemploy_career.head()
```

##Chart 1: Biểu đồ tỷ lệ thất nghiệp và tiếu việc làm từ năm 2018 đến 2022

```python
import plotly.graph_objs as go
years = [2018, 2019, 2020, 2021, 2022]
unemployment={}
underemployed={}
for year in years:
  filt=combined_region[combined_region['Năm'] == year]
  unemployment[year]=filt['Tỷ lệ thất nghiệp'].mean()
  underemployed[year]=filt['Tỷ lệ thiếu việc làm'].mean()
# Create traces for unemployment and underemployment rates
trace_unemployment = go.Scatter(x=years, y=list(unemployment.values()), mode='lines', name='Tỷ lệ thất nghiệp')
trace_underemployed = go.Scatter(x=years, y=list(underemployed.values()), mode='lines', name='Tỷ lệ thiếu việc làm')

# Create layout
layout = go.Layout(
    title='Tỷ lệ thất nghiệp và thiếu việc làm theo năm (2018-2022)',
    xaxis=dict(title='Năm'),
    yaxis=dict(title='Tỷ lệ (%)'))

# Combine traces and layout into a figure
fig = go.Figure(data=[trace_unemployment, trace_underemployed], layout=layout)
fig.update_layout(width=800, height=500)
# Show interactive plot
fig.show()
```

Biểu đồ cho thấy rằng tỷ lệ thất nghiệp và thiếu việc làm tại Việt Nam đã tăng mạnh trong những năm 2020-2022.
* 2020-2021: Đại dịch COVID-19 gây ra gián đoạn kinh tế nghiêm trọng, dẫn đến nhiều doanh nghiệp đóng cửa, thu hẹp hoạt động, và cắt giảm nhân sự. Điều này dẫn đến gia tăng tỷ lệ thất nghiệp và thiếu việc làm trong nhiều ngành nghề.

* Mặc dù tỷ lệ thất nghiệp và thiếu việc làm vẫn cao hơn so với năm 2018 và 2019, nhưng có một dấu hiệu giảm trong năm 2022. Điều này có thể cho thấy sự phục hồi sau đại dịch và các biện pháp khôi phục kinh tế đã bắt đầu có hiệu quả, dẫn đến việc giảm bớt áp lực trên thị trường lao động.Sự sụt giảm vào năm 2022 cho thấy sự phục hồi dần dần, nhưng tỷ lệ vẫn cao do những ảnh hưởng dai dẳng của đại dịch và những thách thức kinh tế đang diễn ra.


## Chart 2: Tỷ lệ các yếu tố ảnh hưởng đến thất nghiệp

```python
un_gen=combined_gender['Tỷ lệ thất nghiệp'].sum()
un_age=combined_age['Tỷ lệ thất nghiệp'].sum()
un_academic=combined_academic['Tỷ lệ thất nghiệp'].sum()
labels = ['Giới tính', 'Tuổi', 'Học vấn']

# Tổng số liệu thất nghiệp của từng nhóm
sizes = [un_gen, un_age, un_academic]

# Màu sắc của các phần tử trong biểu đồ
colors = ['gold', 'lightskyblue', 'lightcoral']

# Vẽ biểu đồ tròn
plt.figure(figsize=(18, 10))
plt.subplot(221)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Đảm bảo biểu đồ tròn
plt.title('Tỷ lệ các yếu tố ảnh hưởng đến tình trạng thất nghiệp')


plt.subplot(222)
# Lặp qua từng trình độ học vấn để vẽ biểu đồ cho mỗi trình độ
for hoc_van in combined_academic['Học vấn'].unique():
    # Lọc dữ liệu cho trình độ học vấn hiện tại
    data = combined_academic[combined_academic['Học vấn'] == hoc_van]


    # Vẽ biểu đồ đường cho trình độ học vấn hiện tại (thiếu việc làm)
    plt.plot(data['Năm'], data['Tỷ lệ thất nghiệp'], label= hoc_van)

plt.title('Xu hướng thất nghiệp theo trình độ học vấn qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)


# giới tính
plt.subplot(223)
plt.plot(combined_gender[combined_gender['Giới tính'] == 'Nam']['Năm'],
         combined_gender[combined_gender['Giới tính'] == 'Nam']['Tỷ lệ thất nghiệp'],
         label='Nam')

plt.plot(combined_gender[combined_gender['Giới tính'] == 'Nữ']['Năm'],
         combined_gender[combined_gender['Giới tính'] == 'Nữ']['Tỷ lệ thất nghiệp'],
         label='Nữ')

plt.title('Tỷ lệ thất nghiệp theo giới tính qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ')
plt.legend()


# tuổi
plt.subplot(224)
for nhom_tuoi in combined_age['Nhóm tuổi'].unique():
    data = combined_age[combined_age['Nhóm tuổi'] == nhom_tuoi]
    plt.plot(data['Năm'], data['Tỷ lệ thất nghiệp'], label=nhom_tuoi)

plt.title('Xu hướng thất nghiệp theo nhóm tuổi qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ thất nghiệp')
plt.legend()
plt.show()

plt.subplots_adjust(wspace=0.2, hspace=0.4)
```

##Chart 3: Biểu đồ tỷ suất sinh và tỷ lệ thất nghiệp (Nữ) theo năm

## Tỷ lệ thất nghiệp/ thiếu việc làm của nam nữ trong mỗi vùng


```python
# Years to visualize
years = [2018, 2019, 2020, 2021, 2022]

for year in years:
    # Filter data for the year
    filtered_data = combined_region[combined_region['Năm'] == year]

    # Create a figure and axes object
    fig, ax = plt.subplots(figsize=(8, 6))

    # Group data by 'Vùng' and 'Giới tính', and calculate the mean unemployment rate
    grouped_data = filtered_data.groupby(['Vùng', 'Giới tính'])['Tỷ lệ thất nghiệp'].mean().unstack()

    # Plot stacked bar plot
    ax = grouped_data.plot(kind='bar', stacked=True, ax=ax, color=['teal', 'orange'])

    plt.title(f'Tỷ lệ thất nghiệp theo phân theo vùng và giới tính trong năm {year}')
    plt.xlabel('Vùng')
    plt.ylabel('Tỷ lệ(%)')
    plt.xticks(rotation=20, ha='right')
    plt.yticks(fontsize=12)

    plt.legend(title='Giới tính')

    # Display percentages on top of each bar
    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy()
        ax.annotate('{:.2f}%'.format(height), (x + width/2, y + height/2), ha='center', va='center')
        plt.ylim(0, 10)

plt.show()
```
